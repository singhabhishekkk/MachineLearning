# Matplotlib example workflow


```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig , axes = plt.subplots(figsize = (5,5)) 
#NOTE : Repeated Mistake  >>>>plt.subplot"s" ...here s is important , if not included surely an error prompts
##figsize = (width, height) of the figure
#Create a figure and a set of subplots

# 3. Plot data 
axes.plot(x,y) ##adding data to our axes

#4. Customize plot
axes.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5.Save & show (you save the whole figure)
fig.savefig("DemoPlot1.png")

plt.show()
```

```python
#Make some data for Histogram and plot it
x = np.random.randn(1000)#Return a sample (or samples) from the "standard normal" distribution.
#Normal Distribution  = Kind of like symmetrical curve
fig,ax = plt.subplots()
ax.hist(x);
#Histogram needs only one axis(ones set of values to plot)
#Here is the main difference between bar charts and histograms. With bar charts,
#each column represents a group defined by a categorical variable; 
#and with histograms, each column represents a group defined by a continuous, quantitative variable.
```

# Two optional for subplots(More than one plot in a cell)

```python
#Subplot option 1
fig,((ax1,ax2),(ax3,ax4))  = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize = (10,5))
#nrows = number of rows of subplot grid
#ncols = number of column of subplot grid
#figsize = (width , height of each figure/plot)

#PLot to each different axis
ax1.plot(x,x/2);#top -left
ax2.scatter(np.random.random(10),np.random.random(10));#top-left
ax3.bar(nutt.keys(),nutt.values());#bottom-right
ax4 . hist(np.random.randn(1000));#bottom-left
```

```python
#Subplot Option 2 :
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = (10,5))

#plot to each different index
ax[0,0].plot(x,x/2);#top -left
ax[0,1].scatter(np.random.random(10),np.random.random(10));#top-left
ax[1,0].bar(nutt.keys(),nutt.values());#bottom-right
ax[1,1].hist(np.random.randn(1000));#bottom-left
#ax[row , columns]
```

# Data visualization Tip
This is a common mistake with beginners.

That is we assume that a graph is always telling the truth.

And we jump to conclusions right away from data.

Instead we want to pause and reflect and make sure that whatever we're visualizing is accurate and testing

it with different methods trying different visualizations.

We don't want to jump to conclusions right away from data.

It takes a lot to understand this data.

So just a quick note that is just because we have data doesn't mean we have knowledge.

# Ideal potting method:
    *pyplot vs matplotlib oo(object oriented) method
    *when plotting something quickly okay to use pyplot method
    *plotting something more advance use oo method

```python
heart
```

```python
#Analysis of patient over 50
over50 = heart[heart["age"]>50]
over50
```

```python
#pyplot method
over50.plot.scatter("age", "chol", c = "target");
#c = The color of each point.
#this colour bar beside the plot is beacuse of c
```

```python
#oo method mixed with pyplot method
fig , ax = plt.subplots(figsize = (10,6))
over50.plot.scatter("age", "chol", c = "target", ax = ax);
#ax = ax indicated to plot the given data in the fig , ax figure
#we get x column also in this plot

#ax.set_xlim([45,100]); #Set the x-axis view limits.
```

```python
## OO method from scratch
fig ,ax = plt.subplots(figsize = (10,6))

#PLot Data
scatter = ax.scatter(over50["age"] , over50["chol"],c = over50["target"]);

#Customize
ax.set(title = "Heart Disease and Cholesterol Levels" , xlabel = "Age" , ylabel = "Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(),title = "Target");#Place a legend on the axes.
#*scatter.legend_elements() = grab the values from scatter and unpack the legend elements 
#name the title of legend to "Target"


#Adding Horizontal line across the axis
ax.axhline(over50["chol"].mean() , linestyle = "--");
```

```python
over50.head()
#thalach = max heart rate
```

```python
#Subplot of chol , age , thalach

fig ,(ax0 ,ax1) = plt.subplots(nrows = 2 , ncols = 1 , figsize = (10,10))


#add data to axes
scatter = ax0.scatter(over50["age"] , over50["chol"],c = over50["target"])

#customization
ax0.set(title = "Heart Disease and Cholesterol Levels" , xlabel = "Age" , ylabel = "Cholesterol");

#Legend
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal line
ax0.axhline(over50["chol"].mean() , linestyle = "--");


#Adding data in ax1
scatter1 = ax1.scatter(over50["age"] , over50["thalach"] , c = over50["target"])

#customizing ax1
ax1.set(title = "Heart Disease and Maximum Heart beat Level" , xlabel = "Age" , ylabel = "Thalach");

#Legend
ax1.legend(*scatter1.legend_elements() , title = "Target");

#horizonal Lines
ax1.axhline(over50["thalach"].mean() , linestyle =  "--" );
#we say here that we need to have the line showing mean value of the thalach column
```

```python
# since the x column is same we can also share the x axis

#Subplot of chol , age , thalach

fig ,(ax0 ,ax1) = plt.subplots(nrows = 2 , ncols = 1 , figsize = (10,10),sharex = True) ########  sharex = True


#add data to axes
scatter = ax0.scatter(over50["age"] , over50["chol"],c = over50["target"])

#customization
ax0.set(title = "Heart Disease and Cholesterol Levels" , xlabel = "Age" , ylabel = "Cholesterol");

#Legend
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal line
ax0.axhline(over50["chol"].mean() , linestyle = "--");


#Adding data in ax1
scatter1 = ax1.scatter(over50["age"] , over50["thalach"] , c = over50["target"])

#customizing ax1
ax1.set(title = "Heart Disease and Maximum Heart beat Level" , xlabel = "Age" , ylabel = "Thalach");

#Legend
ax1.legend(*scatter1.legend_elements() , title = "Target");

#horizonal Lines
ax1.axhline(over50["thalach"].mean() , linestyle =  "--" );
#we say here that we need to have the line showing mean value of the thalach column



#add a title to the figure
fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight = "bold" );
```

```python
# changing style within the existing style


plt.style.use("seaborn-whitegrid")






#Subplot of chol , age , thalach

fig ,(ax0 ,ax1) = plt.subplots(nrows = 2 , ncols = 1 , figsize = (10,10))


#add data to axes
scatter = ax0.scatter(over50["age"] , over50["chol"],c = over50["target"] , cmap = "winter")
#cmap = changes the colour scheme / ColourMap 

#customization
ax0.set(title = "Heart Disease and Cholesterol Levels" , xlabel = "Age" , ylabel = "Cholesterol");

#Legend
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal line
ax0.axhline(over50["chol"].mean() , linestyle = "--");


#Adding data in ax1
scatter1 = ax1.scatter(over50["age"] , over50["thalach"] , c = over50["target"])

#customizing ax1
ax1.set(title = "Heart Disease and Maximum Heart beat Level" , xlabel = "Age" , ylabel = "Thalach");

#Legend
ax1.legend(*scatter1.legend_elements() , title = "Target");

#horizonal Lines
ax1.axhline(over50["thalach"].mean() , linestyle =  "--" );
#we say here that we need to have the line showing mean value of the thalach column
```

```python
#constrain our plot to certain width and height

#Subplot of chol , age , thalach

fig ,(ax0 ,ax1) = plt.subplots(nrows = 2 , ncols = 1 , figsize = (10,10) , sharex = True)


#add data to axes
scatter = ax0.scatter(over50["age"] , over50["chol"],c = over50["target"] ,cmap = "winter")

#customization
ax0.set(title = "Heart Disease and Cholesterol Levels" , xlabel = "Age" , ylabel = "Cholesterol");
ax0.set_xlim([50,80])
#change the x axis limits



#Legend
ax0.legend(*scatter.legend_elements(), title = "Target")

#Horizontal line
ax0.axhline(over50["chol"].mean() , linestyle = "--");


#Adding data in ax1
scatter1 = ax1.scatter(over50["age"] , over50["thalach"] , c = over50["target"] , cmap = "winter")

#customizing ax1
ax1.set(title = "Heart Disease and Maximum Heart beat Level" , xlabel = "Age" , ylabel = "Thalach");
ax1.set_ylim([60,200])


#Legend
ax1.legend(*scatter1.legend_elements() , title = "Target");

#horizonal Lines
ax1.axhline(over50["thalach"].mean() , linestyle =  "--" );
#we say here that we need to have the line showing mean value of the thalach column
```
