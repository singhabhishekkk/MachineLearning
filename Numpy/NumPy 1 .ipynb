## Real World Dot Product Use Case

```python
np.random.seed(0)
#Numbers Of Jar Sold
sales = np.random.randint(20,size = (5,3))
sales,sales.shape
```

```python
#Creating a Weekly Sales DataFrame
#just adding row and column names
weekly = pd.DataFrame(sales,
                      index = ["Mon","Tues","Wed","Thur","Fri"],
                      columns = ["Almond","Peanut","Cashew" ])
weekly
#index = column heading
```

```python
#Pricing of each product
price = np.array([10,8,12])
price,price.shape
#numpy stores prices as 
#10
#8
#12 and not
# 10 8 12
```

```python
#prices dataFrame
pricesData = pd.DataFrame(price.reshape(1,3),
                         index = ["Price"],
                         columns= ["Almond","Peanut","Cashew"])
pricesData
```

```python
#we want to find the total money earned in a day
#so we have to calculate it by dot product
# eg MON  = 12 almond 15 peanut and zero cashew
# total sales on monday = 12 * 10 + 15 * 9 + 0* 15
#there fore we had to do a dot product here
Total = price.dot(sales.T)
Total
#here we multiplied matrices of numbers and we got an array


##To add Total to an DataFrame we need to Find Dot Product of Two DataFrames
```

```python
Totalnew =pricesData.dot(weekly.T) #dot product of two DataFrame
Totalnew
```

```python
type(weekly)
```

```python
type(pricesData)
```

```python
transposed_Totalnew = Totalnew.T 
transposed_Totalnew 
##Transpose of a Pandas DataFrame
```

```python
weekly["Total"] = transposed_Totalnew
```

```python
weekly
```

```python
##we can also use this way of adding a new column rather than the hetic process
weekly["Total"]  = Total
```

```python
weekly
```
